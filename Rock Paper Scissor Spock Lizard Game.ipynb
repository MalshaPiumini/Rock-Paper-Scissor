```python
#Simple Rock Paper Sissor Game

#Rock - R  Paper - P Scissors - C  Spock - S Lizard - L
import random  #generate random values
import getpass #get hidden values
import sys

#Get Input from the user and Convert it into a numeic value
def getInput(choice):
    try:
        valid_choice = choice.upper()
        if valid_choice == 'R':
            return "Rock"
        elif valid_choice == 'P':
            return "Paper"
        elif valid_choice == 'C':
            return "Sissor"
        elif valid_choice =='S':
            return "Spock"
        elif valid_choice =='L':
            return "Lizard"
        else:
            print('Input must be R,P,C,S or L')
            return
    except ValueError:
        val = int(choice)
        print('Input must be R,P,C,S or L')
        return
    
#play with computer
def comPlay():
    print('Computer processing its choice from R,P,C,S,L')
    choices_array = ['Rock','Paper','Sissor','Spock','Lizard']
    random_num = random.randint(0,4) #both numbers include while generating the numbers
    computers_choice = choices_array[random_num]
    player_01 = input('Enter your name :')
    print('Computer choosed his sign, Now Your Turn,')
    choice_with_com =selectChoice(player_01)
    print('Player 1 = Computer')
    print('Player 2 = You')
    findWinner(computers_choice,choice_with_com)
    playAgain = input("If you wanna play again, input Y, else input N")
    try:
        playAgain = playAgain.upper()
        if playAgain == 'Y':
            sys.exit()
        else:
            _main_()
    except ValueError:
        print('Input invalid')
        sys.exit()
    

#play with a friend
def playWithFriend():
    player_1 = input('Enter the name of the player 1 :')
    choice_1 = selectChoice(player_1)
    player_2 = input('Enter the name of the player 2 :')
    choice_2 = selectChoice(player_2)
    final =findWinner(choice_1,choice_2)
    print(final)
    print()
    playAgain = input("If you wanna play again, input Y, else input N")
    try:
        playAgain = playAgain.upper()
        if playAgain == 'Y':
            sys.exit()
        else:
            _main_()
    except ValueError:
        print('Input invalid')
        sys.exit()
        
#select choices
def selectChoice(player):
    print('''Select your choice from these 
             Rock - R
             Paper - P 
             Scissors - C  
             Spock - S 
             Lizard - L ''')
    choice_ = getpass.getpass('Enter the choice of',player)
    try:
        choice_ =choice_.upper()
        if (choice_ != 'R') & (choice_ !='C') & (choice_ !='S') & (choice_ !='L') & (choice_ !='P'):
            print('Your choice is wrong')
            selectChoice(player)
        else:
            return choice_
    except ValueError:
        print('Input must be R,C,S,P or L')
        selectChoice()
    
#Find winner
def findWinner(choice_1,choice_2):
    if choice_1 == choice_2:
        return 'Tie Game'
    
    #Rock and Paper
    elif (choice_1== 'Rock') & (choice_2== 'Paper'):
        return 'Player 2 wins'
    elif (choice_2== 'Rock') & (choice_1== 'Paper'):
        return 'Player 1 wins'
    
    elif (choice_1 == 'Rock') & (choice_2=='Spock'):
        return 'Player 2 wins'
    elif (choice_2 == 'Rock') & (choice_1=='Spock'):
        return 'Player 1 wins'
    
    #Rock and Lizard
    elif (choice_1 == 'Rock') & (choice_2=='Lizard'):
        return 'Player 1 wins'
    elif (choice_2 == 'Rock') & (choice_1=='Lizard'):
        return 'Player 2 wins'
    
    #Paper and Spock
    elif (choice_1 == 'Paper') & (choice_2=='Spock'):
        return 'Player 1 wins'
    elif (choice_1 == 'Spock') & (choice_2=='Paper'):
        return 'Player 2 wins'
    
    #Spock and Lizard
    elif (choice_1 == 'Spock') & (choice_2=='Lizard'):
        return 'Player 2 wins'
    elif (choice_1 == 'Lizard') & (choice_2=='Spock'):
        return 'Player 1 wins'
    
    #Paper and Lizard
    elif (choice_1 == 'Paper') & (choice_2=='Lizard'):
        return 'Player 2 wins'
    elif (choice_2 == 'Paper') & (choice_1=='Lizard'):
        return 'Player 1 wins'
    
    #Paper and scissor
    elif (choice_1 == 'Paper') & (choice_2=='Scissor'):
        return 'Player 2 wins'
    elif (choice_1 == 'Scissor') & (choice_1=='Paper'):
        return 'Player 1 wins'
    
    #Scissor and Rock
    elif (choice_1 == 'Scissor') & (choice_2=='Rock'):
        return 'Player 2 wins'
    elif (choice_1 == 'Rock') & (choice_2=='Scissor'):
        return 'Player 1 wins'
    
    #Scissor and spock
    elif (choice_1 == 'Scissor') & (choice_2=='Spock'):
        return 'Player 2 wins'
    elif (choice_1 == 'Spock') & (choice_2=='Scissor'):
        return 'Player 1 wins'
    
    #Scissor and Lizard
    elif (choice_1 == 'Scissor') & (choice_2=='Lizard'):
        return 'Player 1 wins'
    elif (choice_1 == 'Lizard') & (choice_2=='Scissor'):
        return 'Player 2 wins'
    
    
#Main program
def _main_():
    print('---Welcome to Rock Paper Scissor Spock and Lizard game---')
    print('Select your game :')
    print('    1: Play with a Friend') 
    print('    2: Play with Computer')
    print('    3: Exit')
    ans = input('Enter your choice :')
    
    try:
        user_ans=int(ans)
        if user_ans == 1 :
            playWithFriend()
        elif user_ans ==2:
            comPlay()
        elif user_ans ==3:
            return
        else:
            print('Your input is invalid')
            _main_()
            
    except ValueError:
        print('Input must be a integer')
        _main_()
        
        
#Run Main
_main_() 
```
